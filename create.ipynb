<p itemprop="description">Общими требованиями при оформлении списка литературы проектной работы являются:</p>
<ol>
    <li><s>Единообразие – источники, указанные в списке, должны быть оформлены одинаково.</s></li>
    <li>Выходные данные литературных источников должны позволять находить указанные источники, например, если в списке литературы будет указана книга, оформленная так: «Сулейманов Р.Ф. Музыка в нашей жизни» найти данную книгу будет невозможно, т.к. книга может быть переиздана или быть издана под редакцией, что не указано, а возможно это не книга, а диссертация или статья в журнале. Именно по этой причине необходимо указывать минимальный набор выходных данных, при этом для каждого типа литературы он разнится.</li>
    <li>Источники в списке литературы нумеруются арабскими цифрами с точкой.</li>
    <li>При указании авторов фамилии приводятся в именительном падеже, имя и отчество указываются инициалами.</li>
    <li>Каждая структурная часть в списке литературы отделяется знаком тире (–), знак тире ставится перед:</li>
    <ul>
        <li>местом издания;</li>
        <li>годом издания;</li>
        <li>номером издания;</li>
        <li>страницами (в книгах – перед общим количеством страниц, в журналах – перед страницами, на которых   статья);</li>
        <li>обозначением ссылки на интернет-источник.</li>
    </ul>
    <li>Иностранные источники выносятся в конец списка литературы и оформляются по требованиям, предъявляемым к литературе на русском языке.</li>
</ol>

<h2 id="oform3" class="user-tb-20 wdgt-tl2" itemprop="headline">Оформление нормативных актов (законов, постановлений, указов и т.д.)</h2>
<p itemprop="description">К нормативным актам относятся все законодательные акты, действующие не только на территории Российской Федерации, но и международные акты (как действующие, так и недействующие).</p>
<p itemprop="description">Все законодательные акты имеют одинаковую структуру оформления:</p>
<ol>
	<li>Наименование нормативного акта (Федеральный закон, Приказ, Указ и т.д. (начнем оформлять различные законодательные акты):</li>
		<ul class="user-pb-20 lt-blg-inf-lst">
			<li>Федеральный закон</li>
			<li>Приказ</li>
			<li>Закон</li>
			<li>Кодекс</li>
			<li>Указ Президента</li>
			<li>Постановление Правительства</li>
			<li>Конституция</li>
		</ul>
</ol>

```python
templStart = [
    "Федеральный\sзакон\s",
    "Приказ\s",
    "Закон\s",
    "[А-Я][а-я]+\sкодекс\s",
    "Указ\sПрезидента\s",
    "Постановление\sПравительства\s",
    "Конституция\s"
]
```

<li>Кем был издан нормативный акт (Закон Краснодарского края, Приказ Минспорта России и т.д.), при этом в федеральных законах указывать это не нужно:</li>
		<ul class="user-pb-20 lt-blg-inf-lst">
			<li>Федеральный закон (ни чего указывать не нужно);</li>
			<li>Приказ Минспорта России;</li>
			<li>Закон Краснодарского края.</li>
		</ul>

```python
templFrom = [
    "Мин[а-я]+\sРоссии\s",
    "Российской\sФедерации\s",
    "[А-Я][а-я]+\s[а-я]+\s"
    #Тут могло быть описание всех возможных республик, краёв, автономных округов и т.п. Но через шаблоны с этим будет сложно работать + очень сильно замедлит (поэтому этот пункт не проверяется и считается не обязательным).
]
templFrom = "(" + "|".join(templFrom) + ")"
templFrom
```

<li>Дата, когда был принят нормативный акт, указывается цифрами, где первая цифра – число, далее – месяц, затем – год принятия:</li>
		<ul class="user-pb-20 lt-blg-inf-lst">
			<li>Федеральный закон от 27.07.2004;</li>
			<li>Приказ Минспорта России от 16.03.2017;</li>
			<li>Закон Краснодарского края от 29.12.2020.</li>
		</ul>

```python
templDate = "от\s\d{2}\.\d{2}\.\d{4}(\s)?"
```

<li>Номер нормативного акта. Является уникальным:</li>
	<ul class="user-pb-20 lt-blg-inf-lst">
		<li>Федеральный закон от 27.07.2004 № 79-ФЗ;</li>
		<li>Приказ Минспорта России от 16.03.2017 № 183;</li>
		<li>Закон Краснодарского края от 29.12.2020 № 4403-КЗ.</li>
	</ul>

```python
templNum = "№\s\d{1,}(\-[А-Я]{1,})?(\s)?"
```

<li>Редакция нормативного акта, редакция есть не у всех нормативных актов, ее наличие оформляется в круглых скобках, после открытия скобки указывается «ред. от» и дата редакции (цифрами):</li>
		<ul class="user-pb-20 lt-blg-inf-lst">
			<li>Федеральный закон от 27.07.2004 № 79-ФЗ (ред. от 08.12.2020);</li>
			<li>Приказ Минспорта России от 16.03.2017 № 183 (редакция отсутствует, писать ничего не нужно);</li>
			<li>Закон Краснодарского края от 29.12.2020 № 4403-КЗ (редакция отсутствует, писать ничего не нужно).</li>
		</ul>

```python
templRed = f"(?<=\()ред.\s{templDate}.*?(?=\))"
templRed
```

	<ul class="user-pb-20 lt-blg-inf-lst">

		<li>Приказ Минспорта России от 16.03.2017 № 183 «О признании и включении во Всероссийский реестр видов спорта спортивных дисциплин, видов спорта и внесении изменений во Всероссийский реестр видов спорта»;</li>
