<p itemprop="description">Общими требованиями при оформлении списка литературы проектной работы являются:</p>
<ol>
    <li><s>Единообразие – источники, указанные в списке, должны быть оформлены одинаково.</s></li>
    <li>Выходные данные литературных источников должны позволять находить указанные источники, например, если в списке литературы будет указана книга, оформленная так: «Сулейманов Р.Ф. Музыка в нашей жизни» найти данную книгу будет невозможно, т.к. книга может быть переиздана или быть издана под редакцией, что не указано, а возможно это не книга, а диссертация или статья в журнале. Именно по этой причине необходимо указывать минимальный набор выходных данных, при этом для каждого типа литературы он разнится.</li>
    <li>Источники в списке литературы нумеруются арабскими цифрами с точкой.</li>
    <li>При указании авторов фамилии приводятся в именительном падеже, имя и отчество указываются инициалами.</li>
    <li>Каждая структурная часть в списке литературы отделяется знаком тире (–), знак тире ставится перед:</li>
    <ul>
        <li>местом издания;</li>
        <li>годом издания;</li>
        <li>номером издания;</li>
        <li>страницами (в книгах – перед общим количеством страниц, в журналах – перед страницами, на которых   статья);</li>
        <li>обозначением ссылки на интернет-источник.</li>
    </ul>
    <li><s>Иностранные источники выносятся в конец списка литературы и оформляются по требованиям, предъявляемым к литературе на русском языке.</s></li><i>(иностранные источники будут пропускаться)</i>
</ol>

<h2 id="oform3" class="user-tb-20 wdgt-tl2" itemprop="headline">Оформление нормативных актов (законов, постановлений, указов и т.д.)</h2>
<p itemprop="description">К нормативным актам относятся все законодательные акты, действующие не только на территории Российской Федерации, но и международные акты (как действующие, так и недействующие).</p>
<p itemprop="description">Все законодательные акты имеют одинаковую структуру оформления:</p>
<ol>
	<li>Наименование нормативного акта (Федеральный закон, Приказ, Указ и т.д. (начнем оформлять различные законодательные акты):</li>
		<ul class="user-pb-20 lt-blg-inf-lst">
			<li>Федеральный закон</li>
			<li>Приказ</li>
			<li>Закон</li>
			<li>Кодекс</li>
			<li>Указ Президента</li>
			<li>Постановление Правительства</li>
			<li>Конституция</li>
		</ul>
</ol>

```python
templStart = [
    "Федеральный\sзакон\s",
    "Приказ\s",
    "Закон\s",
    "[А-Я][а-я]+\sкодекс\s",
    "Указ\sПрезидента\s",
    "Постановление\sПравительства\s",
    "Конституция\s"
]
```

<li>Кем был издан нормативный акт (Закон Краснодарского края, Приказ Минспорта России и т.д.), при этом в федеральных законах указывать это не нужно:</li>

```python
templFrom = [
    "Мин[а-я]+\sРоссии\s",
    "Российской\sФедерации\s",
    "[А-Я][а-я]+\s[а-я]+\s"
    #Тут могло быть описание всех возможных республик, краёв, автономных округов и т.п. Но через шаблоны с этим будет сложно работать + очень сильно замедлит (поэтому этот пункт не проверяется и считается не обязательным).
]
templFrom = "(" + "|".join(templFrom) + ")"
```

<li>Дата, когда был принят нормативный акт, указывается цифрами, где первая цифра – число, далее – месяц, затем – год принятия:</li>

```python
templDate = "от\s\d{2}\.\d{2}\.\d{4}(\s)?"
```

<li>Номер нормативного акта. Является уникальным:</li>
	<ul class="user-pb-20 lt-blg-inf-lst">

```python
templNum = "№\s\d{1,}(\-[А-Я]{1,})?(\s)?"
```

<li>Редакция нормативного акта, редакция есть не у всех нормативных актов, ее наличие оформляется в круглых скобках, после открытия скобки указывается «ред. от» и дата редакции (цифрами):</li>
		<ul class="user-pb-20 lt-blg-inf-lst">

<h2 id="oform4" class="user-tb-20 wdgt-tl2" itemprop="headline">Оформление книг (учебные пособия, учебники, монографии)</h2>
<p itemprop="description">При оформлении книг, таких как учебники, монографии, учебные пособия, необходимо указать определенные характеристики книги, по которым, при необходимости, можно найти указанную книгу:</p>

1. Фамилия имя и отчество автора (авторов), при этом фамилия указывается полностью, имя и отчество – инициаперечисляются через запятую. Если авторов более 3-х, то отмечают первых 3-х, затем пишется «и др.».

```python
templOneAuthor = "[А-Я][а-я]+(\s|\,)[А-Я]\.\s?[А-Я]\.\s?"
templSecondAuthor = templOneAuthor*2
templThirdAuthor = templOneAuthor*3 + "(и\sдр\.\s?)?"
templAuthor = templOneAuthor + '(' + templSecondAuthor + ')?(' + templThirdAuthor + ')?'
# Выше в лоб, ниже меньше памяти занимает, с потерей максимум 2-ух фамилий которые мы и так не сохраняем

templAuthor = "[А-Я][а-я]+(\s|\,)[А-Я]\.\s?[А-Я]\.\s?(и\sдр\.\s?)?"
```

2. Название книги – пишется полностью, с первой прописной, без кавычек.

```python
templTitle = "[А-Я].*?[\.\:/]"
```

3. Тип книги – после названия книги ставится двоеточие (:) и указывается тип книги:
<ul>
	<li>учебник;</li>
	<li>монография;</li>
	<li>учебное пособие (в списке литературы сокращается: учеб. пособие).</li>
</ul>

```python
templType = [
    'учебник',
    'монография',
    'учеб\.\sпособие',
    'учебник'
]
templType = [ i + '\.' for i in templType]
temp = '('
for i in templType:
    temp += '(' + i + ')|'
templType = '\s?' + temp[0:-1] + ')'
templType
```

4. Место издания (город), перед которым ставится тире (–), некоторые города необходимо сокращать:
	<ul>
		<li>Москва (М.);</li>
		<li>Ленинград (Л.);</li>
		<li>Санкт-Петербург (СПб.);</li>
		<li>Нижний Новгород (Н. Новгород);</li>
		<li>Ростов-на-Дону (Ростов н/Д.).</li>
	</ul>
<p itemprop="description">Более никакие города не сокращаются, их необходимо писать полностью.</p>

```python
templCity = [
    'М\.',
    'Л\.',
    'СПб\.',
    'Н\.\sНовгород',
    'Ростов\sн\/Д/.',
    '[А-Я].*?[\:\s]'
]

temp = '('
for i in templCity:
    temp += '(' + i + ')|'
templCity = '\s?[\-–]\s?' + temp[0:-1] + ')'
templCity
```

5. Издательство – после города, в котором была издана книга, ставится двоеточие и указывается издательство, оно пишется с большой буквы, по возможности, название издательства сокращается до аббревиатуры, например:
<ul>
    <li>Кубанский государственный университет (КубГУ);</li>
    <li>Агентство международных исследований (АМИ).</li>
</ul>

```python
templIz = '\:?\s[А-Я].*\,\s'
```

6. Год издания – после издательства ставится запятая, указывается год, когда была издана книга.

```python
templAge = '\d{4}\.'
```

7. Количество страниц – после года ставится точка, затем тире и указывается общее количество страниц в книге, затем ставится буква «с.», обозначающая количество страниц.

```python
templPage = '\s[\-–]\s?((\d{1,}\sс)|(С\.second\d{1,}))\.'
```

<p itemprop="description">Дополнительными элементами при оформлении книги могут быть:</p>
<ul>
    <s><li>том (указывается после названия книги, если есть переводчик или книга издана под редакцией, то после данных элементов);</li></s>
    <li>переводчик (указывается после названия книги, например, / Пер. с англ. О.П. Бокова, инициалы переводчика указываются перед фамилией);</li> 
    <li>под редакцией (указывается после названия книги, например, / Под ред. И.В. Светловой, инициалы переводчика указываются перед фамилией).</li>
</ul>

```python
templRedBook = '/?\s((Под\sред\.)|(Пер\.\sc\s[а-я]+))\s?'             
templAuthorRe = '[А-Я]\.[А-Я]\.\s[А-Я][а-я]+\.'
```

<ul class="user-pb-20 lt-blg-inf-lst">
    <li>Фамилия, имя, отчество автора (авторов) – перечисляются через запятую.</li>
    <li>Название статьи – записывается с заглавной буквы без кавычек.</li>
    <li>Две косых черты (//) являются разделителем между заголовком статьи и названием журнала (Название статьи // Название журнала).</li>
    <li>Название журнала – указывается полностью, с заглавной буквы, в кавычки не берется.</li>
    <li>Год издания – после названия журнала ставится точка и тире, далее указывается год, когда был издан журнал, из которого был взят материал.</li>

Страница, как и в сборнике, обозначается заглавной буквой «С», после которой ставится точка и указываются страницы, на которых расположена статья, номера страниц записываются через тире.</li>
